```python
import open3d as o3d
import matplotlib.pyplot as plt
import numpy as np
from open3d import *
import pandas as pd
```

```python
mesh = o3d.io.read_point_cloud("kota_circuit2.ply")
```

```python
data = np.asarray(mesh.normals)
point = np.asarray(mesh.points)
color = np.asarray(mesh.colors)
```

```python
df = pd.DataFrame({'x':point[:,0],'y':point[:,1],'z':point[:,2], 'nx':data[:,0],'ny':data[:,1], \
                   'nz':data[:,2],'red':color[:,0] ,'green':color[:,1],'blue':color[:,2]})
```

```python
df #  z>=.9, -0.2<x<+0.2, -0.2<y<+0.2
```

```python
cut_out = df[((-0.005 < df['nx']) & (df['nx'] < 0.005)) & ((-0.02 < df['ny']) & (df['ny'] < 0.02)) & ( df['nz'] > 0.93)]
```

```python
cut_out
```

```python
pcd = o3d.geometry.PointCloud()
```

```python
pcd.points = o3d.utility.Vector3dVector(cut_out[['x','y','z']].values)
```

```python
pcd.normals = o3d.utility.Vector3dVector(cut_out[['nx','ny','nz']].values)
```

```python
pcd.colors = o3d.utility.Vector3dVector(cut_out[['red','green','blue']].values)
```

```python
o3d.visualization.draw_geometries([pcd])
```
